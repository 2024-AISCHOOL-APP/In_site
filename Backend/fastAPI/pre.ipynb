```python
import pandas as pd
import pickle

# 기존 데이터를 불러옵니다.
dress_data = pd.read_pickle('dress_data_with_tokens.pkl')

dress_data['prod_name'].unique()

# '스크릿웨딩드레스'를 가진 행 찾기
target_row = dress_data[dress_data['prod_name'] == '스크릿웨딩드레스']

# 행의 인덱스 가져오기
target_index = target_row.index

# 행의 이름(인덱스) 변경하기
dress_data.rename(index={target_index[0]: '시크릿웨딩드레스'}, inplace=True)

store_idx_values = {
    '스크릿웨딩드레스': 22,
    'W웨딩': 23,
    '베일스': 42,
    '라미엔브라이드': 25,
    '브라이덜헤미쉬': 29
}

# 조건에 맞는 경우 store_idx 컬럼을 추가하고 값을 할당
dress_data['prod_idx'] = dress_data['prod_name'].map(store_idx_values)
dress_data.isnull().sum()

dress_data.to_pickle('dress_data_with_tokens_updated.pkl')
```
